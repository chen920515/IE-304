```python
#Poisson distribution
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Define the Poisson distribution parameter lambda
lam = 4
```

```python
# Create an array of x values
x = np.arange(0, 12).tolist()
```

```python
# Create the Poisson probability mass function (PMF)
pmf = poisson.pmf(x, lam)
```

```python
# Create the plot
plt.bar(x, pmf)
plt.title('Poisson Probability Mass Function', fontsize=16)
plt.xlabel('Number of events', fontsize=16)
plt.ylabel('Probability', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
