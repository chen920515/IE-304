```python
# Five-Point Summary
df.boxplot(column="Score1", fontsize=20)
plt.show()
df.boxplot(column="Weight", by="Smokes", fontsize=20)
plt.show()
fig = plt.figure(figsize=(16, 8))
plt.subplots_adjust(top=0.8, right=0.5, hspace=5, wspace=3)
df.hist(layout=(8,3))
plt.show()
df["Weight"].hist(bins=10)
plt.show()
```

```python
#SNS
#increase font size of all elements
sns.set(font_scale=1.5)
sns.boxplot(x="Score1",data=df).set(title='Boxplot')
plt.show()
#plot multiple lines
sns.lineplot(x='Date', y='Sales', hue='Company', data=df).set(title='Sales Data')
plt.show()
```
